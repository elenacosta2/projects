# Data Selection and Setup

## FAO 'Food Balance Sheets' Dataset

The Food and Agriculture Organization of the United Nations (FAO) provides free access to food and agriculture data. The dataset used in this analysis was the  **'Food Balance Sheets' dataset**, which reports the pattern of 174 countries' **food supply** from 1961 to 2013 (latest update), measured as 1000 tonnes units.

As reported in the [dataset description](https://www.fao.org/faostat/en/#data/FBSH), the yearly food supply reported in the dataset was calculated by adding the total quantity of foodstuffs produced in a country added to the total quantity imported and adjusted to any change in stocks that may have occurred since the beginning of the reference period.

The dataset is focused on two utilizations of each food item available:
+ **Food** - refers to the total amount of the food item available as human food during the reference period.
+ **Feed** - refers to the quantity of the food item available for feeding to the livestock and poultry during the reference period.

## 'Environmental Impact of Food Production' Dataset

The **'Environment Impact of Food Production'** dataset contains 43 common foods grown across the globe and 23 columns as their respective land, water usage and carbon footprints. As stated in the [dataset FAQs](https://ourworldindata.org/faqs-environmental-impacts-food), these represent **greenhouse gas (GHG) emissions** expressed as **kg of CO2 equivalents per kg of food product** across different **stages in the lifecycle** of food production:
+ *Land use change* 
+ *Animal Feed production*
+ *On Farm* impacts in crop or livestock production (including the manufacturing of inputs such as fertilizers, or emissions from manure)
+ *Food Processing*: the conversion of raw ingredients into sold products, such as the processing of cereals into bread
+ *Transport*: this includes transport from the farm up to retail. Transport of food from retail to consumers’ homes is not included.
+ *Packaging*
+ *Retail*: energy consumption in retail stores, such as refrigeration.

Kg of CO2 equivalents is a unit of measure that takes into account all types of GHG using global warming potential (over 100-year timescale), a value which aims to represent the amount of warming that each specific gas generates relative to CO2. For example, 1kg of methane (CH4) corresponds to 28kg of CO2 (IPCC) in terms of global warming potential. 

In addition, GHG emissions and other environmental impacts are reported related to the food's **mass, kilocalories and protein content**:
- *Eutrophying emissions (gPO₄eq)*: eutrophication refers the pollution of a superficial water body due to the addition of minerals and nutrients such as nitrogen and phosphorous. This excessive organic load can lead to the complete death of the ecosystem, since it promotes the excessive growth of algae which consume all the oxygen, causing all aerobic organisms to die.

- *Freshwater withdrawals (liters)*: 'Water withdrawals [...] are defined as freshwater taken from ground or surface water sources, either permanently or temporarily, and conveyed to a place of use.' ([OECD](https://data.oecd.org/water/water-withdrawals.htm#:~:text=Definition%20of,to%20a%20place%20of%20use.))

- *Land use (m²)*: 'Land use involves the management and modification of natural environment or wilderness into built environment such as settlements and semi-natural habitats such as arable fields, pastures, and managed woods.' ([Wikipedia](https://en.wikipedia.org/wiki/Land_use))

- *'Scarcity-weighted water use (liters)* represents freshwater use weighted by local water scarcity, combining water use and water stress in one indicator (however, several research papers do not recomment this indicator).

### Regions

For later analysis, it could be useful to group countries according to region ('Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America'). To do this, we will use the pycountry_convert library by saving its directory in the current directory.

First, we must check that Area Abbreviation codes correspond to correct alpha3 country codes. 

```python
# import library containing conversion dictionaries between codes, country names and continents
from pycountry_convert import country_alpha3_to_country_alpha2 as alpha
from pycountry_convert import country_alpha2_to_continent as cont

# get the dictionary containing alpha3 codes
codes_dict = alpha.COUNTRY_ALPHA3_TO_COUNTRY_ALPHA2

# get a set of all alpha3 codes
alpha3 = set([i for i in codes_dict.keys()])

# get a set of all Area Abbreviations in our dataset
alpha_FAO = set(df_FAO['Area Abbreviation'])

# get Area Abbreviations that are not included in the alpha3 codes
alpha_FAO - alpha3
```

The only missing country in the library is Timor Leste, so we will add it manually.

```python
# define a function to go from alpha3 to alpha2 to continent
def continent_gen(country_alpha3):
    if country_alpha3 == 'TLS':
        country_continent = 'Asia'
    else: 
        country_alpha2 = alpha.convert_country_alpha3_to_country_alpha2(country_alpha3) # from alpha3 to alpha2
        country_continent = cont.convert_country_alpha2_to_continent(country_alpha2) # from alpha2 to continent
    return country_continent

# insert new column 'Region' defining the continent
df_FAO.insert(3, 'Region', df_FAO['Area Abbreviation'].apply(lambda x: continent_gen(x)))

# remove Area Abbreviation
df_FAO.drop(['Area Abbreviation'], axis=1, inplace=True)
```

However, if we look at data from 1992 to 2013, the **global food supply increased at a higher rate than the population** and if the trends are maintained in future years, the spread will continue to increase.

We know that global population increase will likely slow down in future years, but this may not come with a slow down in food production, as developing countries will also improve their lifestyles.

## Per Capite Food Supply

To compare different countries, it is important to analyse the total food supply in relation to the country's population.  

The only contry in the FAO dataset that is not present in the population dataset is Taiwan, therefore we will create a new FAO dataset by omitting this country to make this analysis.

```python
# identify index of rows to drop
idx_drop = df_FAO[df_FAO.Area == 'China, Taiwan Province of'].index

# remove the rows
df_FAO_noT = df_FAO.drop(idx_drop, axis=0)
```

We will then create a new dataframe to store information about pro capite food supply in every country in 2013, and plot the countries with the highest and lowest pro capite food supply for easier visualization. All top 10 countries in 2013 will be included and labelled with the color red. 

```python
# create dataframe with pro capite supply in 2013
d = []

for c in sorted(set(df_FAO_noT.Area)):
    y_supply = df_FAO_noT[(df_FAO_noT.Area == c) & (df_FAO_noT.Year == 2013)]['Supply'].sum()
    c_pop = df_pop[(df_pop.Country == c) & (df_pop.Year == 2013)]['Population'].sum() # by making an agg func i don't return the index
    d.append([c, y_supply/c_pop*1000])
    
df_pc13 = pd.DataFrame(d, columns=('Country', 'Supply'))
df_pc13.set_index('Country', inplace=True) # set index
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

# map colors to largest producers in 2013
df_pc13['largest'] = np.where(df_pc13.index.isin(c_largest13.index), 'tab:red', 'tab:blue')
df_ordered = df_pc13[(df_pc13.Supply > 2.5) | (df_pc13.Supply < 1) | (df_pc13.index.isin(c_largest13.index))][['Supply', 'largest']].sort_values('Supply')
colors = df_ordered['largest']

# plot countries with very low or high pro capite food supply, highlighting largest producers (all included)
df_pc13[(df_pc13.Supply > 2.5) | (df_pc13.Supply < 1) | (df_pc13.index.isin(c_largest13.index))]['Supply'].sort_values().plot(kind='barh', 
                                                                                                                            color=colors.values)
ax.set_xlabel('Per capite Food Supply (tonne/person)')
ax.set_title('Countries ranked for per capite food supply (2013)')

plt.show()
```

```python
for country in ['United States of America','China, mainland','India']:
    a = df_pc13.loc[country, 'Supply']
    print('Food supply in {} in 2013 was {:.2f} tonnes per capite.'.format(country, a))
```

We can see that among the top food producers, **The United States of America** have the highest pro-capite food supply. On the other hand, **India**, which is the second country for food supply, has a very low pro capite food supply, less than 1 tonne per person. Moreover, none of the top producers are among the ones with the highest food supply per capite: these countries are **Denmark, Belarus, Ireland, Serbia, Canada and Ukraine**, which all go beyond 3 tonnes of food per capite.
